```python
import numpy as np
import pandas as pd
import pylab as pl
%pylab inline
```

```python
mymean = 100
```

```python
distributions = ['normal', 'poisson', 'chisquare', 'lognormal', 'binomial']
```

```python
import random
random.seed(99)
size = {}
all_sample_means = {}


for i in range(len(distributions)):
    random.seed(i)
    size[distributions[i]] = random.sample(range(10, 2000),100)
    print('The number of samples of ' + distributions[i] + ' distribution is ' + str(len(size[distributions[i]])))
    all_sample_means[distributions[i] + '_sample_means'] = []

print('all_sample_means: ' + str(all_sample_means))
```

```python
size_df = pd.DataFrame(size)
size_df.describe()
```

```python
normal_samples = {}
for x in size['normal']:
    normal_samples[x] = np.random.normal(loc = mymean, scale = 10, size = x)

fig1 = pl.figure(figsize=(10,6)).add_subplot(111)
normal_samples_means = []
for n in normal_samples:
    sample_mean = normal_samples[n].mean()
    all_sample_means['normal_sample_means'].append(sample_mean)
    fig1.plot(n, sample_mean, 'o')
    fig1.plot([min(size['normal']), max(size['normal'])], [mymean, mymean], 'k')
    fig1.set_xlabel('sample size', fontsize=18)
    fig1.set_ylabel('sample mean', fontsize=18)
    fig1.set_title('normal ' + 'distribution', fontsize=18)
    #print(sample_mean)
figtext(.01, .02, 'This plot describes normal distribution in terms of Central Limit Theorem', fontsize = 12, ha = 'center')
   
```

```python
poisson_samples = {}
for x in size['poisson']:
    poisson_samples[x] = np.random.poisson(lam = mymean, size = x)

fig2 = pl.figure(figsize=(10,6)).add_subplot(111)
for n in poisson_samples:
    sample_mean = poisson_samples[n].mean()
    all_sample_means['poisson_sample_means'].append(sample_mean)
    fig2.plot(n, sample_mean, 'o')
    fig2.plot([min(size['poisson']), max(size['poisson'])], [mymean, mymean], 'k')
    fig2.set_xlabel('sample size', fontsize=18)
    fig2.set_ylabel('sample mean', fontsize=18)
    fig2.set_title('poisson ' + 'distribution', fontsize=18)
    #print(sample_mean)
figtext(.01, .02, 'This plot describes poisson distribution in terms of Central Limit Theorem', fontsize = 12, ha = 'center')
```

```python
chisquare_samples = {}
for x in size['chisquare']:
    chisquare_samples[x] = np.random.chisquare(df = mymean, size = x)

fig3 = pl.figure(figsize=(10,6)).add_subplot(111)
for n in chisquare_samples:
    sample_mean = chisquare_samples[n].mean()
    all_sample_means['chisquare_sample_means'].append(sample_mean)
    fig3.plot(n, sample_mean, 'o')
    fig3.plot([min(size['chisquare']), max(size['chisquare'])], [mymean, mymean], 'k')
    fig3.set_xlabel('sample size', fontsize=18)
    fig3.set_ylabel('sample mean', fontsize=18)
    fig3.set_title('Chi-squared ' + 'distribution', fontsize=18)
    #print(sample_mean)
figtext(.01, .02, 'This plot describes chisquare distribution in terms of Central Limit Theorem', fontsize = 12, ha = 'center')
```

```python
lognormal_samples = {}
for x in size['lognormal']:
    lognormal_samples[x] = np.random.lognormal(mean = np.log(mymean) - 1/2, sigma = 1, size = x)

fig4 = pl.figure(figsize=(10,6)).add_subplot(111)
for n in lognormal_samples:
    sample_mean = lognormal_samples[n].mean()
    all_sample_means['lognormal_sample_means'].append(sample_mean)
    fig4.plot(n, sample_mean, 'o')
    fig4.plot([min(size['lognormal']), max(size['lognormal'])], [mymean, mymean], 'k')
    fig4.set_xlabel('sample size', fontsize=18)
    fig4.set_ylabel('sample mean', fontsize=18)
    fig4.set_title('lognormal ' + 'distribution', fontsize=18)
    #print(sample_mean)
figtext(.01, .02, 'This plot describes lognormal distribution in terms of Central Limit Theorem', fontsize = 12, ha = 'center')
```

```python
binomial_samples = {}
for x in size['binomial']:
    binomial_samples[x] = np.random.binomial(n = mymean/0.5, p = 0.5, size = x)
fig5 = pl.figure(figsize=(10,6)).add_subplot(111)
for n in binomial_samples:
    sample_mean = binomial_samples[n].mean()
    all_sample_means['binomial_sample_means'].append(sample_mean)
    fig5.plot(n, sample_mean, 'o')
    fig5.plot([min(size['binomial']), max(size['binomial'])], [mymean, mymean], 'k')
    fig5.set_xlabel('sample size', fontsize=18)
    fig5.set_ylabel('sample mean', fontsize=18)
    fig5.set_title('binormal ' + 'distribution', fontsize= 18)
    
    #print(sample_mean)
figtext(.01, .02, 'This plot describes binominal distribution in terms of Central Limit Theorem', fontsize = 12, ha = 'center')

```

```python
fig6 = pl.figure(figsize=(10,6)).add_subplot(211)
fig6.set_xlabel('sample mean', fontsize = 18)
fig6.set_ylabel('N', fontsize = 18)
fig6.set_title('All sample means of 5 distributions', fontsize = 18)

fig = {}
n = 0
for k, v in all_sample_means.items():
    n += 1    
    fig[n] = fig6.hist(v, bins = 30, label = k)
fig6.legend()

fig7 = pl.figure(figsize=(10,6)).add_subplot(212)
fig7.set_xlabel('sample mean', fontsize = 18)
fig7.set_ylabel('N', fontsize = 18)
fig7.set_title('All sample means except lognormal distributions', fontsize = 18)
klist = list(all_sample_means.keys())
fig7.hist(all_sample_means[klist[0]], bins = 30, label = klist[0])
fig7.hist(all_sample_means[klist[1]], bins = 30, label = klist[1])
fig7.hist(all_sample_means[klist[2]], bins = 30, label = klist[2])
fig7.hist(all_sample_means[klist[4]], bins = 30, label = klist[4])
fig7.legend()

figtext(.01, .02, 'The plots show that lognormal distribution is less effective than other 4 distributions in terms of Central Limit Theorem', fontsize = 12, ha = 'center')
```
